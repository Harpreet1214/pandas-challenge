```python
import csv
import pandas as pd
```

```python
# Save path to data set in a variable
data_file = "Resources/purchase_data.csv"
```

```python
# Use Pandas to read data
purchase_data = pd.read_csv(data_file)
purchase_data.head()
```

```python
#Calculate total number of players
total = purchase_data["SN"].unique()
pd.DataFrame({"Total Players":[len(total)]})
```

```python
#Purchasing Analysis
# Number of Unique Items
unique_items = len(purchase_data["Item Name"].unique())

#Average Purchase Price
average_purchaseprice = purchase_data["Price"].mean()

#Total Number of Purchases
Total_purchase = purchase_data["Purchase ID"].count()

#Total Revenue
Total_revenue = purchase_data["Price"].sum()

pd.DataFrame({"Number of Unique Items": [unique_items],
                              "Average Price": average_purchaseprice,
                              "Number of Purchases": Total_purchase,
                              "Total Revenue": Total_revenue})
```

```python
#Gender Demographics
#Count of Male, Female and Other/Non-Disclosed Players
Gender_sn = purchase_data.loc[:, ["Gender", "SN"]]
Gender_sn = Gender_sn.drop_duplicates()
count = Gender_sn["Gender"].value_counts()
Gender_count = purchase_data["Gender"].value_counts()


#Percentage of Male, Female and Other/Non-Disclosed Players
percentage = Gender_count/len(purchase_data)*100

pd.DataFrame({"Total Count": count,
              "Percentage of Players":percentage
             
})
```

```python
#Purchasing Analysis (Gender)
 
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
Purchase_count = purchase_data.groupby(["Gender"]).count()["Price"]
Average_Purchaseprice = purchase_data.groupby(["Gender"]).mean()["Price"]
Total_purchase = purchase_data.groupby(["Gender"]).sum()["Price"]
Purchase_totalperperson = Total_purchase/count


#Create a summary data frame to hold the results
summary_dataframe = pd.DataFrame({"Purchase Count": Purchase_count,
              "Average Purchase Price": Average_Purchaseprice,
                "Total Purchase Value": Total_purchase,
                 "Avg Total Purchase per Person": Purchase_totalperperson})


#Display the summary data frame
summary_dataframe
```

```python
#Age Demographics
 
#Establish bins for ages
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999999999]
Group_Names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
Players_age = purchase_data.loc[:,["Gender", "SN", "Age"]]
Players_age = Players_age.drop_duplicates()
Players_count = Players_age.count()[0]


#Categorize the existing players using the age bins. Hint: use pd.cut()
Players_age["Age Range"]= pd.cut(Players_age["Age"], bins, labels = Group_Names) 


#Calculate the numbers and percentages by age group
Players_agecount = Players_age["Age Range"].value_counts()
Players_percentage = Players_agecount/Players_count*100


#Create a summary data frame to hold the results
Age_Demographics_dataframe = pd.DataFrame({"Total Count": Players_agecount,
             "Percentage of Players": Players_percentage})


#Display Age Demographics Table,
Age_Demographics_dataframe.sort_index()
```
